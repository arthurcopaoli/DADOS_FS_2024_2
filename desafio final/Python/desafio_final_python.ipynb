```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/dados_rh.csv')

df
```

```python
linhas, colunas = df.shape

print(f'A tabela tem {linhas} linhas e {colunas} colunas.')
```

```python
dados_null = df.isnull().sum()
print("Colunas com dados nulos e suas quantidades:")
print(dados_null)
```

```python
tipos_dados = df.dtypes
print("Tipos de dados no banco de dados:")
print(tipos_dados)
```

```python
grafico_setores = df['Department'].value_counts()

grafico_setores.plot.bar(color = 'navy', title = 'Quantidade de Funcionarios por Setor')
plt.xlabel('Setor', fontsize=12)
plt.ylabel('Quantidade de Funcionarios', fontsize=12)
plt.xticks(rotation=360)
```

```python
grafico_escolaridade = df['Education'].value_counts().sort_index()

grafico_escolaridade.plot.line(color = 'green', title = 'Quantidade de Funcionarios por Nivel de Escolaridade')
plt.xlabel('Nivel de Escolaridade', fontsize=12)
plt.ylabel('Quantidade de Funcionaarios', fontsize=12)
```

```python
grafico_atrito = df['Attrition'].value_counts()

grafico_atrito.plot.pie(autopct= '%1.1f%%', title = 'Funcionarios Com ou Sem Atrito')
```
